<h3>Data frame</h3>
Is when u import data like excel and panda will turn it into a data frame

```python
#create data frame
#arguments are (data, rows, col, )
rows=["Row1", "Row2", "Row3", "Row4","Row5"]
col=["col1","col2","col3","col4"]
df=pd.DataFrame(np.arange(0,20).reshape(5,4),rows,col)            
```

<h2>Accessing elements</h2>
Basically a data frame is a collection of series. A data frame in our case should have at least 2 rows and at least 2 col. If it has one row only it would be a series

```python
#1. by just accessing the column
df["col1"]
```

```python
#we can also access multiple columns by creating a list
df[["col1","col2"]]
```

```python
#2. using .loc[]
#only row indexes
df.loc["Row1"]
```

```python
#if we check the type we see that it is a series
type(df.loc["Row1"])
```

```python
#3. using .iloc[:,:]
#both row and col indexes
df.iloc[0:2,0:4]
```

```python
#now if we chceck the type it would be data frame
type(df.iloc[0:2,0:4])
```

<h1>CSV</h1>

```python
from io import StringIO, BytesIO
```

```python
#this is a string that we will convert to csv file
data=("col1,col2,col3\n"
     "a,b,1\n"
     "c,d,2\n"
     "x,y,3\n")
```

```python
pd.read_csv(StringIO(data))
```

```python
#read from specific columns
dataF=pd.read_csv(StringIO(data),usecols=lambda x: x.upper() in ["COL1","COL3"])
#we can also write is as
#dataF=pd.read_csv(StringIO(data),usecols=["col1","col3"])
```

```python
dataF
```

```python
#we can convert the data to different data types
#instead of onject we can also use int,float etc
dataF=pd.read_csv(StringIO(data),dtype=object)
```

```python
#as we can see the data type is object
print(dataF['col1'])
#more exactly the type of object is string
dataF['col1'][0]
```

```python
data1=("col1,col2,col3\n"
     "1,2,3\n"
     "4,5,6\n"
     "7,8,9\n")
```

```python
#we can assign different data types to each column
#for dtype we are using a dictonary of key-value pairs
dataF=pd.read_csv(StringIO(data1),dtype={"col1":int,"col2":np.float64,"col3":"Int64"})
```

```python
dataF
```

```python
#we can check the data type of all columns
dataF.dtypes
```

```python
#we can assing an index colmuns to be one of our columns
#the following line will assing col3 (which has index2) as the index column
pd.read_csv(StringIO(data1),index_col=2)
```

```python
#sometimes if the first col is int, it will be automatically coneverted to index col
#to avoid it we can have an argument: index_col=False
```
