```python
import seaborn as sns
```

```python
#import a dataset
df=sns.load_dataset('tips')
```

```python
#visualize data
df.head()
```

```python
# in our case, 'tip' is the dependent feature, the rest of them are independent
```

<h1>Correlation with heatmap</h1>
This way we can see the relation between two idependent features

```python
df.dtypes
```

```python
#here i create a new dataset to contain only float and int
df1 = df.select_dtypes(include=['float64', 'int64'])
df1.corr()
```

```python
#we can also use colors
#we pass the matrix of correlation as the argument
sns.heatmap(df1.corr())
```

<h1>Join plot</h1>

<h3>Univariate analysis</h3>

```python
#the hex parameter gives us hexagons in the graph
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
#the reg parameter gives us regression in the graph
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

<h2>Pair plot</h2>
For more than 2 features.

A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
# with hue we can select one parameter
sns.pairplot(df,hue='sex')
```

<h2>Dist plot</h2>
Helps us check the distributions of columns feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False)
```

<h1>Categorical plots</h1>
Seaborn also allows us to analyse categorical data points

```python
#count plot
#this is for 'sex'
sns.countplot(x='sex',data=df)
```

```python
#this is for 'day'
sns.countplot(x='day',data=df)
```

```python
#we can do that on the y axis too
sns.countplot(y='day',data=df)
```

```python
#bar plot
# 2 categorries plotted with respect to one another
sns.barplot(x='sex',y='total_bill',data=df)
```

<h2>Box plot</h2>

```python
sns.boxplot(x='smoker',y='total_bill',data=df)
```

```python
# use different colors
sns.boxplot(hue='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

<h2>Violin plot</h2>
Helps us see the distribution of elements in terms of kernel density estimation and box plot

```python
sns.violinplot(x='total_bill', hue='day',data=df, palette="rainbow")
```
