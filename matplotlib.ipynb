<h1>Matplotlib</h1>

Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+. 

Some of the major Pros of Matplotlib are : 

• Generally easy to get started for simple plots 

• Support for custom labels and texts 

• Great control of every element in a figure 

• High-quality output in many formats Very customizable in general 1

```python
import matplotlib.pyplot as plt
#this is a shortcut to display graphs
%matplotlib inline
```

```python
import numpy as np
```

```python
#create 2 variables
x=np.arange(0,10)
y=np.arange(11,21)
```

<h3>Scatter</h3>

```python
# here i scatter with color green, by default, scatter(x,y) would be blue
plt.scatter(x,y,c='g')

# if we didn't have the line %matplotlib inline at the start we would need the following line to display the graph:
# plt.show()
```

```python
#add labels and title
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D graph')

#we can also save the resulting image
plt.savefig("first-graph.png")
```

<h3>Plot</h3>
Apart from scatter we are now getting a straight line

```python
#first let s change y and make a polynomial graph
y=x*x
plt.plot(x,y,'r')
```

```python
#if we want to have a graph od dashes we include it in the color formating
plt.plot(x,y,'r--')
```

```python
#this will also add the scatter points
plt.plot(x,y,'ro--')
```

```python
#we can also make dashed line with another parameter 'linestyle'
# we also can thicken the graph 'linewidth'
# as for the '*' charachters for scatter we can shange their size with 'markersize'

plt.plot(x,y,'r*', linestyle='dashed',linewidth=5,markersize=10)
```

<h3>Create subplots</h3>
Within one diagram we can create multiple subplots

```python
#assign the number of rows and columns of the canvas, third paramter is the position
# setting the rows and col as 2,2 will give as the possiblity to have 4 graphs

#first subplot
plt.subplot(2,2,1)
plt.plot(x,y,'go')
#second
plt.subplot(2,2,2)
plt.plot(x,y,'r--')
#third 
plt.subplot(2,2,3)
plt.plot(x,y,'y',linewidth=5)
```

```python
#make a sine graph
#the step paramter(third one) will make the graph smoother if smaller 
w=np.arange(0,6*np.pi,0.1)
z=np.sin(w)

plt.plot(w,z)
```

```python
# short example of sine and cos
a=np.arange(0,4*np.pi,0.1)
b_sin=np.sin(a)
b_cos=np.cos(a)

#making subplots
#if we choose the position to be the same in both subplots, they will be displayed in the same garph
plt.subplot(1,2,1)
plt.plot(a,b_sin)
plt.title("sine wave")
plt.subplot(1,2,1)
plt.plot(a,b_cos)
plt.title("cos wave")
```

<h3>Bar plot</h3>

```python
x1 = [2,8,10] 
y1 = [11,16,9]  

plt.bar(x1, y1) 
```

```python
x2 = [3,9,11] 
y2 = [6,15,7] 
plt.bar(x2, y2, color = 'g') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis')  
```

```python
#merge the graphs
plt.bar(x1, y1) 
plt.bar(x2, y2, color = 'g') 
```

<h3>Histograms</h3>

```python
#this whows the denisty of numbers in intervals
#by default the interval is 10, but we can change it "bins"
c=np.array([1,2,3,4,5,13,15,33,34,35,36,37,38,39,42,55,56])
plt.hist(c,bins=20)
```

<h3>Box plots</h3>

```python
data=[np.random.normal(0,std,100) for std in range (1,4)]
# 2nd parameter is vert. is we set it to false the plot will be horizontal
# 3rd paramter is color if set True, otherwise we'll have no color
plt.boxplot(data,vert=True, patch_artist=True)
```

<h3>Pie chart</h3>

```python
labels='men', 'women', 'elders', 'children'
size=[40,60,50,10]
#this swhow how further from the graph would be one category displayed/moved
explode=(0,0.1,0,0)
colors=['gold', 'yellow','yellowgreen','lightblue']

# autopct is the type of percentaeg we want to be displayed(float/int)
plt.pie(size, explode=explode,labels=labels, colors=colors, autopct='%i%%', shadow=True)
plt.title("Population")
```
